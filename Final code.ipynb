```python
import requests 
user_input = int(input("Would you like to search by ISBN or textbook name? Press 1 for ISBN, press 2 for textbook name."))
if user_input == 1: 
    isbn = input("Please enter your ISBN number:")
    r = requests.get('https://www.googleapis.com/books/v1/volumes?q=isbn:'+ isbn)
elif user_input == 2:
    text_book_name = input("Please enter the name of the textbook you are looking for : ")
    r = requests.get('https://www.googleapis.com/books/v1/volumes?q=title:'+ text_book_name)
r = r.json()
list_of_books = r['items']
for book in list_of_books:
    sales_info = book['saleInfo']
    book_info = book['volumeInfo']
    print('Title: '+ book_info['title'])
    authors = book_info['authors']
    print('Authors: ')
    for author in authors:
        print(author)
    if 'buyLink' in sales_info: 
        print('Link to buy: ')
        print(sales_info['buyLink'])
    else:
        print("This book is not available on google books, we are sorry!")
        print('\n')
        continue 
    if 'listPrice' in sales_info:
        print('Price: ')
        print('$'+ str(sales_info['listPrice']['amount']))
    else: 
        print("The book is free!")
    print('\n')
```
